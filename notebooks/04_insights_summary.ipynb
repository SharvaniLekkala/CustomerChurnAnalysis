```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/processed/churn_featured.csv")
```

```python
os.makedirs("../outputs/plots", exist_ok=True)
```

```python
def save_and_show_plot(filename):
    """Saves the current matplotlib figure and shows it."""
    plt.tight_layout()
    plt.savefig(f"../outputs/plots/{filename}", dpi=300)
    plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x="churn", data=df, palette="pastel")
plt.title("Churn Distribution")
plt.ylabel("Number of Customers")
save_and_show_plot("churn_distribution.png")
```

```python
df['churn_numeric'] = df['churn'].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="contract", y="churn_numeric", data=df, estimator="mean", palette="pastel")
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.xlabel("Contract Type")
save_and_show_plot("churn_by_contract.png")
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="tenure", hue="churn", multiple="stack", bins=30, palette="pastel")
plt.title("Churn Distribution by Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
save_and_show_plot("churn_by_tenure.png")
```

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(data=df, x="monthlycharges", hue="churn", fill=True, common_norm=False, palette="pastel")
plt.title("Churn Distribution by Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Density")
save_and_show_plot("churn_by_monthlycharges.png")
```
