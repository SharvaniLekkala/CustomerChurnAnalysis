```python
import pandas as pd

df = pd.read_csv("../data/processed/churn_cleaned.csv")
df.head()
```

```python
def tenure_category(tenure):
    if tenure <= 12:
        return "0-1 year"
    elif tenure <= 24:
        return "1-2 years"
    elif tenure <= 48:
        return "2-4 years"
    else:
        return "4+ years"

df["tenure_category"] = df["tenure"].apply(tenure_category)
```

```python
df["charges_category"] = pd.cut(df["monthlycharges"],
                                bins=[0, 30, 60, 90, 150],
                                labels=["Low", "Medium", "High", "Very High"])
```

```python
df["avg_charge_per_month"] = df["totalcharges"] / df["tenure"]
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/churn_featured.csv", index=False)
print("Feature-engineered data saved")
```
